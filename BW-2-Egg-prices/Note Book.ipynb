```python
import pandas as pd 
import numpy as np
```

```python
year = 2019
dframes = []
for i in range(6):
    dframes.append(pd.read_html(f'https://marketnews.usda.gov/mnp/py-report?&repDate=01%2F01%2F{year}&regionsDesc=&endDate=12%2F31%2F{year}&repMonth=1&run=Run&_producttypefrom=1&endYear=2023&producttypefrom=&repYear=2020&categoryDesc=Egg&frequency=Daily&datatype=None+Selected&previouscls=Breaking+Stock&_producttype=1&report=NW_PY041&category=Egg&subcategory=Breaking+Stock&endMonth=1&commodityDesc=Breaking+Stock&runReport=true&format=excel' ))
    year += 1
```

```python
# Question 1 : create a single data frame.
df = dframes[0][0]
for i in range(6):
    df = pd.concat([df, dframes[i][0]])
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date',inplace=True)
df
```

```python
# Question 2 : What was the average low price for eggs in each of the years of data?
df.resample('YS')['Low Price'].mean()
```

```python
# Question 3 : In what year was the percentage increase the greatest?
df.resample('YS')['Low Price'].mean().pct_change().dropna().apply(lambda x:x*100).apply(lambda x:np.int64(x))
```

```python
# Question 4 : get the average low price for eggs in each month and year of data.
df['Low Price'].resample('MS').mean().head(50)
```

```python
# Question 5 : Have egg prices gone down in the last few months?
df['Low Price'].resample('MS').mean().sort_index(ascending=False).head(6)
```
